```python
def get_frequency(input_string):

    list_of_words = input_string.split(' ')
    dict_of_frequencies = {}

    for word in list_of_words:

        if word in dict_of_frequencies.keys():
            dict_of_frequencies[word] = dict_of_frequencies[word] + 1
        else:
            dict_of_frequencies[word] = 1

    return(max(dict_of_frequencies.values())/len(list_of_words))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('/Users/bryannicholson/Downloads/email.csv')

data['hyperlink'] = None
data['free'] = None
data['click'] = None
data['business'] = None
data['frequency'] = None

email_data=data['email']
key_spam_words=['hyperlink','free','click','business']

for email in email_data:
    for word in key_spam_words:
        if word in str(email):
            data[word]=1
        data['frequency']=get_frequency(str(email))
        
word_frequency=pd.DataFrame([data['hyperlink'],data['free'],data['click'],data['business'],data['frequency']])

labels=pd.DataFrame([data['label']])

x_train, x_test, y_train, y_test=train_test_split(word_frequency.T,np.ravel(labels.T),test_size=0.3)

logistic_model=LogisticRegression()
fit_model=logistic_model.fit(x_train,y_train)

y_pred=fit_model.predict(x_test)

print('Accuracy:', accuracy_score(y_test,y_pred))
```
