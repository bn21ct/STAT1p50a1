```python
from random import sample

for i in range(0,100):

    a=sample(['1','2','3',],1)
    b=sample(['1','2','3',],1)
    
    file_a=open(file='file1.txt',mode='a')
    file_b=open(file='file2.txt',mode='a')
    
    file_a.write('\n'+a[0])
    file_b.write('\n'+b[0])
    
    file_a.close()
    file_b.close()
    
with open ('file1.txt','r') as f1, open ('file2.txt','r') as f2:
    lines_a = f1.readlines()
    lines_b = f2.readlines()
    
    for line_a in lines_a:
        for line_b in lines_b:
            if line_a == line_b:
                print("Success")
            else:
                pass
        
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
data=pd.read_csv('/Users/bryannicholson/Downloads/crime.csv')

data=data[['PctPopUnderPov','ViolentCrimesPerPop']]

crime_list = np.array(data['ViolentCrimesPerPop']).reshape(-1,1)
poverty_list = np.array(data['PctPopUnderPov']).reshape(-1,1)

crime_scaled=StandardScaler().fit_transform(crime_list)
poverty_scaled=StandardScaler().fit_transform(poverty_list)

plt.scatter(data['PctPopUnderPov'],data['ViolentCrimesPerPop'])
plt.xlabel('PctPopUnderPov  - Scaled')
plt.ylabel('ViolentCrimesPerPop - Scaled')
plt.legend()
plt.show()

plt.scatter(crime_scaled,poverty_scaled)
plt.xlabel('PctPopUnderPov')
plt.ylabel('ViolentCrimesPerPop')
plt.legend()
plt.show()
```

```python
import statistics
def OutlierFinder(listOfValues,testValue):
    mean=statistics.mean(listOfValues)
    sdv=statistics.stdev(listOfValues)
    upLimit=mean-sdv*3
    lowLimit=mean+sdv*3
    if testValue>upLimit or testValue< lowLimit:
        print('The test value is an outlier')
```

```python
OutlierFinder([1,2,3,4,5],33)
```
